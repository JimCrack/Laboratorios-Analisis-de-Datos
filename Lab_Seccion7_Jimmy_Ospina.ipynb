1. Crear un array de números del 1 al 10 y mostrar la suma, la media, y el valor máximo.

```python
import numpy as np

num= np.arange(1,11)
print(num)
suma = num.sum()
print("Suma:",suma)
print("Media:",num.mean())
print("Máximo:",num.max())
```

2. Crear un array de 3x4 lleno de ceros y luego cambiar su forma a 2x6

```python
import numpy as np
zeros=np.zeros((3,4))  #Creamos la matriz de ceros de 3x4
print(zeros)

print()

arr = zeros.reshape(2,6) #Cambiamos su forma a matriz de 2x6
print(arr)
```

3. Generar una matriz aleatoria de 3x3 y calcular su transpuesta.

```python
import numpy as np

matrice = np.random.rand(3,3)  # Crear una matriz 3x3 con valores aleatorios
print("Matriz original")
print(matrice)

print()

matrice_transpuesta= matrice.T   # Calcula la matriz transpuesta
print("Matriz Transpuesta")
print(matrice_transpuesta)
```

4.	Concatenar dos arrays de NumPy, cada uno con 5 valores, y luego imprimir el array concatenado.

```python
x = np.random.rand(1,5)
y = np.random.rand(1,5)
print("Array a:",x)
print()
print("Array b:",y)
print()

arr_concatenado = np.concatenate((x,y))
print("Array final:",arr_concatenado)
```
